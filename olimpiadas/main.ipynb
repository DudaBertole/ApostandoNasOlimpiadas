## Tratamento
Me interessa apenas 
- O "ano" da participação do atleta
- A "equipe" do atleta 
- O "esporte" da participação
- A "medalha" conquitada pelo atleta

```python
columns_to_keep = ['ano', 'equipe', 'esporte', 'medalha'] 
```

```python
df_olimp_smp = df_olimp[columns_to_keep].copy()
df_olimp_smp
```

```python
df_olimp_smp = df_olimp_smp[df_olimp_smp['ano'] >= 2000]
df_olimp_smp
```

```python
df_olimp_smp['medalha'].value_counts()
```

```python
df_olimp_smp['esporte'].value_counts()
```

```python
brazilian_athletes = df_olimp_smp[df_olimp_smp['equipe'] == 'Brazil']
brazilian_athletes['esporte'].value_counts()
```

```python
brazil_gymnastics = df_olimp_smp[
    (df_olimp_smp['equipe'] == 'Brazil') & (df_olimp_smp['esporte'] == 'Gymnastics')
]

brazil_gymnastics
```

```python
brazil_gymnastics.to_excel("brazil_gymnastics.xlsx", index=False)
```

```python
brazil_athletics = df_olimp_smp[
    (df_olimp_smp['equipe'] == 'Brazil') & (df_olimp_smp['esporte'] == 'Athletics')
]

brazil_athletics
```

```python
china_gymnastics = df_olimp_smp[
    (df_olimp_smp['equipe'] == 'China') & (df_olimp_smp['esporte'] == 'Gymnastics')
]

china_gymnastics
```

```python
china_gymnastics.to_excel("china_gymnastics.xlsx", index=False)
```
